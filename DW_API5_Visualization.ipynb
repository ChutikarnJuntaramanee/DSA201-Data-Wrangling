```python
#2 Data USA
rawdata2 = requests.get('https://datausa.io/api/data?drilldowns=Nation&measures=Population')
print(rawdata2.json())
```

```python
# Assign new variable
rawdata2_json = rawdata2.content
```

```python
data2_json = json.loads(rawdata2_json)
data2_json
```

```python
data_2 = data2_json['data']
Data2 = pd.DataFrame(data_2)
Data2
```

```python
Data2.drop(['ID Year', 'Slug Nation'],axis ='columns', inplace= True)
Data2
```

```python
Data2.sort_values("Year")
```

```python
Data2.info()
```

```python
data_year_type = {'Year' : int}
Data2 = Data2.astype(data_year_type)
Data2.dtypes
```

```python
#Line graph2 Data USA
import matplotlib.pyplot as plt
Data2.plot(x ='Year', y='Population', figsize = (15,5), xlabel="Year", ylabel="Population", title= "Number of population by year on Data USA")
```

```python
#pie graph
population = Data2['Population']
year = Data2['Year']
explode = (0.2, 0.2, 0, 0, 0, 0, 0)
colors = ['#009392','#39B185','#9CCB86', '#E9E29C','#EEB479','#E88471', '#CF597E']
plt.pie(population, labels=year, autopct='%1.1f%%', startangle=120, shadow = True, colors=colors, explode=explode)
plt.title('Data USA Population till 2019')



```
